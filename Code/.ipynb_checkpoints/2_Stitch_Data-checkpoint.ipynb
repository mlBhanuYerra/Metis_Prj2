```python
def read_file_DataFrame(var_key, tuple_file_info):
    '''
    Function to read the variables from the csv files and create a pandas DataFrame
    
    args: a key for the var, a tuple of (type of var, filename,  date_label, date_format)
    returns: a pandas DataFrame with date in datetime format labeled as "Date"
             and variable column labeled in var_key format
    '''
    
    file_path = '../Data/' + tuple_file_info[1]
    df = pd.read_csv(file_path)
    
    df["Date"]=pd.to_datetime(df[tuple_file_info[2]], format=tuple_file_info[3])
    
    df.rename(columns = lambda col: col.strip(), inplace=True)
    
    df[var_key]=df[tuple_file_info[-1]]
    
    return df.loc[:,["Date", var_key]]

list_of_dataframes_to_stitch = []
for key, tup in data_dict.items():
    list_of_dataframes_to_stitch.append(read_file_DataFrame(key, tup))

print([df.shape for df in list_of_dataframes_to_stitch])
```

```python
CPI_tbl = list_of_dataframes_to_stitch.pop(5)
SP500_tbl = list_of_dataframes_to_stitch.pop(0)
print(len(list_of_dataframes_to_stitch))
```

```python
for i in range(len(list_of_dataframes_to_stitch)):
    SP500_tbl = SP500_tbl.merge(list_of_dataframes_to_stitch[i], on='Date', how='left')
SP500_tbl
```

```python
SP500_tbl["GoldFutures"] = pd.to_numeric(SP500_tbl["GoldFutures"], errors='coerce')
SP500_tbl["OilSpot"] = pd.to_numeric(SP500_tbl["OilSpot"], errors='coerce')
SP500_tbl
```
