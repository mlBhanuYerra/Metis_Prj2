```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import time
import scipy.stats as stats

import seaborn as sns
sns.set()

from sklearn.model_selection import train_test_split
from sklearn.model_selection import TimeSeriesSplit
from sklearn.linear_model import LinearRegression, Lasso, LassoCV, Ridge, RidgeCV
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures

import statsmodels.api as sm

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [15, 5]
```

```python
def R_sqrd(actual_ys, pred_ys):
    actual_ys=pd.Series(actual_ys)
    pred_ys = pd.Series(pred_ys)
    
    sse = np.sum(np.square(actual_ys - pred_ys))
    avg_actuals = np.mean(actual_ys)
    sst = np.sum(np.square(actual_ys-avg_actuals))
    
    return (1 - sse/sst)    
```

```python
actual_ys = pd.Series([1, 2,3,4,5])
resd_ys = pd.Series([1.1, 2.1,3.1,4.1,5.1])

R_sqrd(actual_ys, resd_ys)
```

```python
GP_df = pd.read_csv('../Data/final_Data_for_Modeling.csv', parse_dates=['Date'])
GP_df
```

```python
GP_df['Gold_next'] = GP_df['Gold'].shift(periods=-1)
GP_df['Gold_1week'] = GP_df['Gold'].shift(periods=-5)
GP_df['Gold_4week'] = GP_df['Gold'].shift(periods=-28)
GP_df['Gold_prev'] = GP_df['Gold'].shift(periods=1)
# Gold percentge change (todays over previous close)
GP_df['Gold_chg'] = 100*(GP_df['Gold'] - GP_df['Gold_prev'])/GP_df['Gold_prev']
#
GP_df['GPInc'] = (GP_df['Gold_next']/GP_df['Gold']-1)
GP_df["year"]=pd.to_datetime(GP_df['Date'], format='%Y-%m-%d').dt.year - 2000
GP_df
```

```python
train_set_def = GP_df['year']<=15
test_set_def = (GP_df['year']>15) * (GP_df['year']<=18)
```

```python
def linear_wCoefs(xt, yt, xval, yval):
    lm = LinearRegression()

    lm.fit(xt, yt)
    # score fit model on validation data
    val_score = lm.score(xval, yval)
    
    # report results
    print('\nValidation R^2 score was:', val_score)
    print('Feature coefficient results: \n')
    if lm.fit_intercept:
        print("Intercept : {:.8f}".format(lm.intercept_))
    
    for feature, coef in zip(xt.columns, lm.coef_):
        print(feature, ':', f'{coef:.8f}')
    
    pred = lm.predict(xval)
    
    xvar1 = xt.columns[0]
    #yvar = yt.columns[0]
    
    plt.subplot(1, 3, 1)
    plt.scatter(xval[xvar1],yval)
    #plt.plot(xval['Gold'], pred, color='blue',linewidth=1)
    plt.title("Regression fit")
    plt.xlabel(xvar1)
    plt.ylabel('y')
    
    plt.subplot(1, 3, 2)
    res = yval - pred
    plt.scatter(pred, res)
    plt.title("Residual plot")
    plt.xlabel("prediction")
    plt.ylabel("residuals")
    
    plt.subplot(1, 3, 3)
    #Generates a probability plot of sample data against the quantiles of a 
    # specified theoretical distribution 
    stats.probplot(res, dist="norm", plot=plt)
    plt.title("Normal Q-Q plot")
```

```python
def return_GoldDF(xvars, yvars):
    """
    """
    x_train = GP_df.loc[train_set_def, xvars].dropna(axis=0).copy()
    y_train = GP_df.loc[train_set_def, xvars + [yvars]].dropna(axis=0)[yvars].copy()

    x_test = GP_df.loc[test_set_def, xvars].dropna(axis=0).copy()
    y_test = GP_df.loc[test_set_def, xvars + [yvars]].dropna(axis=0)[yvars].copy()
    
    return x_train, x_test, y_train, y_test
```

## Model 1: Next Day's Forecast with % Increase in Gold

```python
plt.rcParams['figure.figsize'] = [15, 5]
x_vars1 = ['Gold', 'Gold_chg', 'GoldFutures','Silver', 'Platinum', 'Palladium', 'OilSpot', \
             'DJI', 'FTSE', 'TBill', 'FedFundRates', 'USDX']
y_var1 = 'Gold_next'

x_train1, x_test1, y_train1, y_test1 = return_GoldDF(x_vars1, y_var1)

linear_wCoefs(x_train1, y_train1, x_test1, y_test1)
```

```python
fit1 = sm.OLS(y_train1, x_train1).fit()
fit1.summary()
```

## Model 2: Next Day's Forecast without % Increase in Gold

```python
x_vars2 = ['Gold', 'GoldFutures','Silver', 'Platinum', 'Palladium', 'OilSpot', \
             'DJI', 'FTSE', 'TBill', 'FedFundRates', 'USDX']
y_var2 = 'Gold_next'

x_train2, x_test2, y_train2, y_test2 = return_GoldDF(x_vars2, y_var2)

linear_wCoefs(x_train2, y_train2, x_test2, y_test2)
```

## Scaling the Model Inputs

```python
def return_GoldDF_scaled(xvars, yvars):
    """
    """
    scaler = StandardScaler()
    
    
    x_train = pd.DataFrame(scaler.fit_transform(GP_df.loc[train_set_def, xvars].dropna(axis=0)), columns=xvars)
    y_train = GP_df.loc[train_set_def, xvars + [yvars]].dropna(axis=0)[yvars].copy()

    x_test = pd.DataFrame(scaler.transform(GP_df.loc[test_set_def, xvars].dropna(axis=0)), columns=xvars)
    y_test = GP_df.loc[test_set_def, xvars + [yvars]].dropna(axis=0)[yvars].copy()
    
    return x_train, x_test, y_train, y_test
```

```python
plt.rcParams['figure.figsize'] = [15, 5]
x_vars2 = ['Gold', 'GoldFutures','Silver', 'Platinum', 'Palladium', 'OilSpot', \
             'DJI', 'FTSE', 'TBill', 'FedFundRates', 'USDX']
y_var2 = 'Gold_next'

x_train2b, x_test2b, y_train2b, y_test2b = return_GoldDF_scaled(x_vars2, y_var2)

linear_wCoefs(x_train2b, y_train2b, x_test2b, y_test2b)
```

```python
fit2b = sm.OLS(y_train2b, x_train2b).fit()
fit2b.summary()
```

```python
ypred_2b = fit2b.predict(x_train2b)

sse = np.sum(np.square(y_train2b-ypred_2b))

mean_gp = np.mean(y_train2b)
sst = np.sum(np.square(y_train2b-mean_gp))

R_sqrd_trivial = 1 - sse/sst
print(R_sqrd_trivial)
```
