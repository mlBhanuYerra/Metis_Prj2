```python
train_set_def = Gold_pd['year']<=15
test_set_def = (Gold_pd['year']>15) * (Gold_pd['year']<=18)

x_train = Gold_pd.loc[train_set_def, ['Gold', 'Silver', 'Platinum', 'Palladium']].copy()
y_train = Gold_pd.loc[train_set_def, 'Gold_next'].copy()

x_test = Gold_pd.loc[test_set_def, ['Gold', 'Silver', 'Platinum', 'Palladium']].copy()
y_test = Gold_pd.loc[test_set_def, 'Gold_next'].copy()
```

```python
# --- Model
'''
Simple linear regression with gold, silver, platinum and palladium prices
'''
poly = PolynomialFeatures(degree=2) 
x_train_poly = pd.DataFrame(poly.fit_transform(x_train))
x_test_poly = pd.DataFrame(poly.fit_transform(x_test))
linear_wCoefs(x_train_poly, y_train, x_test_poly, y_test)
```

```python
# --- Model 2
'''
Simple linear regression 
without today's gold price but with silver, platinum and palladium prices
'''
x_train2 = Gold_pd.loc[train_set_def, ['Silver', 'Platinum', 'Palladium']].copy()
y_train2 = Gold_pd.loc[train_set_def, 'Gold_next'].copy()

x_test2 = Gold_pd.loc[test_set_def, ['Silver', 'Platinum', 'Palladium']].copy()
y_test2 = Gold_pd.loc[test_set_def, 'Gold_next'].copy()
                                       
linear_wCoefs(x_train2, y_train2, x_test2, y_test2)                  
```

```python
# --- Model 3
'''
Simple linear regression 
without today's gold price but silver, silver **2 and a quadratic in platinum and palladium prices
'''

x_train3 = x_train2.copy()
x_test3 = x_test2.copy()

x_train3['Silver2'] = x_train3['Silver'] ** 2
x_train3['Plat2'] = x_train3['Platinum'] ** 2
x_train3['Pallad2'] = x_train3['Palladium'] ** 2
x_train3['PlatPallad'] = x_train3['Platinum'] * x_train3['Palladium'] 

x_test3['Silver2'] = x_test3['Silver'] ** 2
x_test3['Plat2'] = x_test3['Platinum'] ** 2
x_test3['Pallad2'] = x_test3['Palladium'] ** 2
x_test3['PlatPallad'] = x_test3['Platinum'] * x_test3['Palladium']

linear_wCoefs(x_train3, y_train2, x_test3, y_test2)      
```

```python
# --- Model 4
'''
Simple linear regression 
without today's gold price but silver, silver **2 and a quadratic in platinum and palladium prices
'''

x_train4 = x_train3.copy()
x_test4 = x_test3.copy()

x_train4['Gold'] = x_train['Gold']
x_test4['Gold'] = x_test['Gold']

linear_wCoefs(x_train4, y_train2, x_test4, y_test2)      
```
